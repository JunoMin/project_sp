```
# 토지특성정보(속성) api(국가공간정보센터) 사용하기

def getLandCharacteristics(pnu):

  url = "http://openapi.nsdi.go.kr/nsdi/LandCharacteristicsService/attr/getLandCharacteristics"
  queryParams = '?' + urlencode({quote_plus('authkey'): '27678dcebb1e5b972d22f6'
                               , quote_plus('pnu') : pnu #/* 고유번호(8자리 이상) */  
                               , quote_plus('stdrYear') : '2020' #/* 기준연도(YYYY: 4자리) */  
                               , quote_plus('format') : 'json' #/* 응답결과 형식(xml 또는 json) */  
                               #, quote_plus('numOfRows') : '10' #/* 검색건수 */  
                               #, quote_plus('pageNo') : '1' #/* 페이지 번호 */  
                                })
  # 콤마 주의
  request = Request(url + queryParams)
  request.get_method = lambda: 'GET'
  response_body = urlopen(request).read()
  res = json.loads(response_body.decode('utf-8'))
  df = pd.DataFrame(res['landCharacteristicss']['field'])

  return df
```

```
# 공동주택가격정보(속성) api(국가공간정보센터) 사용
def getApartHousingPriceAttr(pnu):
    url = "http://openapi.nsdi.go.kr/nsdi/ApartHousingPriceService/attr/getApartHousingPriceAttr"
    queryParams = '?' + urlencode({quote_plus('authkey'): '40c4a9b0bcb091e9ac0afa'
                                      , quote_plus('pnu'): pnu  # /* 고유번호(8자리 이상) */
                                      , quote_plus('stdrYear'): '2020'  # /* 기준연도(YYYY: 4자리) */
                                      , quote_plus('format'): 'json'  # /* 응답결과 형식(xml 또는 json) */
                                      , quote_plus('numOfRows'): '10000'  # /* 검색건수 */ 설정하지 않을 경우 최대 10개만 출력
                                   # , quote_plus('pageNo') : '1' #/* 페이지 번호 */
                                   })
    # 콤마 주의
    request = Request(url + queryParams)
    request.get_method = lambda: 'GET'
    response_body = urlopen(request).read()
    res = json.loads(response_body.decode('utf-8'))
    df = pd.DataFrame(res['apartHousingPrices']['field'])
    return df
```
